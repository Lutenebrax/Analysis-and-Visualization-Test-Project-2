For this project we will use a dataset containing the Euro daily exchange rate between 1999 and 2021.

This is a quick project in which we will use a different way to create graphic data visualization related to the Euro-dollar exchange rate during some non-covid 19 pandemic years and the covid pandemic years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from pandas.tseries.offsets import DateOffset
%matplotlib inline

exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
display(exchange_rates.head())
display(exchange_rates.tail())
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_dollar = exchange_rates[["Time", "US_dollar"]]
print(euro_to_dollar["US_dollar"].value_counts())

euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"] != "-"]
print(euro_to_dollar["US_dollar"].value_counts())

euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)

euro_to_dollar.info()
```

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
euro_to_dollar
```

```python
time_years = euro_to_dollar[["Time", "US_dollar"]]
time_2016_to_2019 = time_years[(time_years["Time"] >= dt.datetime(2016, 1, 1)) & (time_years["Time"] <= dt.datetime(2019, 12, 30))]
time_2020 = time_years[(time_years["Time"] >= dt.datetime(2020, 1, 1)) & (time_years["Time"] <= dt.datetime(2020, 12, 30))]

plt.figure(figsize = (10, 20))
plt.subplot(2, 1, 1)
plt.plot(time_2016_to_2019["Time"], time_2016_to_2019["US_dollar"])
plt.xticks(rotation = 45)
plt.ylim([1.03, 1.26])
plt.xlabel("Euro_dollar rate")
plt.ylabel("Years 2016 - 2019")
plt.title("Euro-dollar rate in the non-pandemic(2016-2019) years")
plt.subplot(2, 1, 2)
plt.plot(time_2020["Time"], time_2020["US_dollar"])
plt.xticks(rotation = 45)
plt.ylim([1.03, 1.26])
plt.xlabel("Euro_dollar rate")
plt.ylabel("Year 2020")
plt.title("Euro-dollar rate in the pandemic(2020) year")
plt.show()
```

Here we show the comparison of the euro and dollar rate between the pandemic and non-pandemic years.

```python
time_years = euro_to_dollar[["Time", "rolling_mean"]]
time_2016_to_2019 = time_years[(time_years["Time"] >= dt.datetime(2016, 1, 1)) & (time_years["Time"] <= dt.datetime(2019, 12, 30))]
time_2020 = time_years[(time_years["Time"] >= dt.datetime(2020, 1, 1)) & (time_years["Time"] <= dt.datetime(2020, 12, 30))]

plt.figure(figsize = (10, 20))
plt.subplot(2, 1, 1)
plt.plot(time_2016_to_2019["Time"], time_2016_to_2019["rolling_mean"])
plt.xticks(rotation = 45)
plt.ylim([1.03, 1.26])
plt.xlabel("Euro_dollar rate")
plt.ylabel("Years 2016 - 2019")
plt.title("Euro-dollar rate in the non-pandemic(2016-2019) years")
plt.subplot(2, 1, 2)
plt.plot(time_2020["Time"], time_2020["rolling_mean"])
plt.xticks(rotation = 45)
plt.ylim([1.03, 1.26])
plt.xlabel("Euro_dollar rate")
plt.ylabel("Year 2020")
plt.title("Euro-dollar rate in the pandemic(2020) year")
plt.show()
```

Here we show the same graphs as above but using the rolling().mean() method to display smoother results.

Since we are taking the years from 2016 to 2020 as the Non-covid 19 pandemic years (4 years), is hard to visually compare them to the 1 year worth of data for the Covid 19 pandemic years.
Even then, we can see that for the pandemic years, the Euro-dollar exchange rate has been increasing most part of the time.
